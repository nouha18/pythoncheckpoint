```python
###checkpoint 3
import pandas as pand

col_list = ["PassengerId","Survived","Pclass","Name","Sex","Age","SibSp","Parch","Ticket","Fare","Cabin"]
##df = pand.read_csv("data.csv",sep=";")

df = pand.read_csv("data.csv",sep=";",usecols=col_list)
print("Affichage des valeurs null,NaN \n")
print(df.describe()) 
##ou print(df.info()) 
print(df.head(1461).isnull().sum())
print('this is a sum of nan values of All columns'+'\033[94m')
##PassengerId      0
##Survived         0
##Pclass         891
##Name           891
##Sex            891
##Age              0
##SibSp          891
##Parch          891
##Ticket         891
##Fare           891
##Cabin


print(df["Age"].head(1461).isnull().sum())
print(df["Age"].fillna(0,inplace=True))
print(df["Age"])
```

```python

print(df["Age"].head(1461).isnull().sum())
print(df["Age"].fillna(0,inplace=True))
print(df["Age"])

print('Survived columns ******'+'\033[94m')
##print(df["Survived"])
print(df["Survived"].head(1461).isnull().sum())
print(df["Survived"].fillna("survived",inplace=True))
print('Name columns ******'+'\033[94m')
print(df["Name"])
print(df.head(1461).isnull().sum())
print(df["Name"].fillna("person",inplace=True))

print('Pclass columns ******'+'\033[94m')
print(df["Pclass"])
print(df.head(1461).isnull().sum())
print(df["Pclass"].fillna("underclass",inplace=True))

print('SibSp columns ******'+'\033[94m')
print(df["SibSp"].head(10))
print(df.head(1461).isnull())
print(df["SibSp"].fillna("0",inplace=True))

print('Parch columns ******'+'\033[94m')
print(df["Parch"])
print(df.head(1461).isnull().sum())
print(df["Parch"].fillna("0000",inplace=True))
print(df.head(1461).isnull().sum())
print('genre columns ******'+'\033[94m')
print(df["Sex"].head(10))
print(df.head(1461).isnull())
print(df["Sex"].fillna("male",inplace=True))

print('Ticket columns ******'+'\033[94m')
print(df["Ticket"])
print(df.head(1461).isnull().sum())
print(df["Ticket"].fillna("Second",inplace=True))

print('Fare columns ******'+'\033[94m')
print(df["Fare"])
print(df.head(1461).isnull().sum())
print(df["Fare"].fillna("Second",inplace=True))

print('Cabin columns ******'+'\033[94m')
print(df["Cabin"])
print(df["Cabin"].head(1461).isnull().sum())
print(df["Cabin"].fillna("vip",inplace=True))
print('Fare columns ******'+'\033[94m')
##df.drop('Fare', axis=1, inplace=True)
print("nombre des valeurs null")

final_df = df
print(final_df.head(1461).isnull())
##export as dataframe with "," separator
file = final_df.to_csv('tooldata.csv')





```

```python

import seaborn as sns
import pandas as pand
import numpy as np
import matplotlib.pyplot as plt
dfile = pand.read_csv("titanic.csv")
print("Correlation result ")
###########################
def plot_correlation_map( dfile ):

        corr = dfile.corr()
        sns.set_theme(style="white")
        # Set up the matplotlib figure
        s , ax = plt.subplots( figsize =( 11 , 9 ) )
        cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

        s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
        plt.show()
        
        
print(plot_correlation_map( dfile ))

            
```

**Description of Plot:** 
Dans ce diagramme on remarque que la variation des couleurs varie du rose vif vers le bleu intense du plus intense vers le moins intense.
plus la relation est intense entre les deux paramètres et plus la couleurs tend vers le rose donc vers la valeur=1
si non elle tend vers le beu de valeur null ou négative.
